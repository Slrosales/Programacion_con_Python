## Librerías

Es un conjunto de módulos o funciones enfocados en facilitar la resolución de problemas en áreas específicas.

Partes de una líberia:


*   **Documentación:** describe como usar las funciones y clases.
*   **Módulos:** archivos de código que contienen funciones, clases y varibales.
*   **Funciones y clases:** diseñadas para resolver problemas especificos.
*   **Dependencias:** Algunas libreías pueden depender de otras para funcionar
*   **Importación:** Puedes importar la libreía completa o ciertas funciones en concreto

## **Arrays**


Un array es una estructura de datos que organiza datos del mismo tipo en una matriz multidimensional. Son fundamentales para realizar operaciones matemáticas y manipulaciones en el área de análisis y ciencia de datos.

Funciones básicas que se pueden aplicar a los arrays:

```python
# Crear un vector lleno de ceros de 8 elementos
V = np.zeros((8))
print(V, "\n")

# Crear una matriz (2 dimensiones) llena de unos
# 3 filas, 2 columnas
M = np.ones((3,2))
print(M)
```

```python
# Crear un array vacío de 5 espacios
array_vacio = np.empty(5)

# Valores a insertar en el array
valores = [1, 3, 2, 4, 5]
lista = []

for valor in valores:
    lista.append(valor)

# Agregar el número 5
array_vacio = np.append(array_vacio, 5)

# Insertar en el indice 2 el número 3
array_vacio[2] = 3

# Imprimir el array
print(array_vacio)
```

```python
# Crear un array de una dimensión manualmente
vector = np.array([1, 3, 2, 4, 5, 5])
```

```python
print("Dimensión del vector: ",vector.shape,"\n")   # Es un array unidimensional que contiene 6 elementos
print("El elemento en el indice 2 es: ",vector[2],"\n")      # Acceder al tercer elemento (índice 2)
print(f"{vector} + 10 = {vector +10}\n")     # Sumar 10 a cada elemento del array
```

```python
# Estadísticas y Cálculos Numéricos

print(f"Promedio: {np.mean(vector)}")
print(f"Mediana: {np.median(vector)}")
print(f"Desviación estándar: {np.std(vector)}")
print(f"Varianza: {np.var(vector)}\n")

print(f"Máximo valor en el vector: {np.max(vector)}")
print(f"Mínimo valor en el vector: {np.min(vector)}\n")

print(f"La suma de los elementos: {np.sum(vector)}")     # Suma todos los elementos del array.
print(f"La suma acumulativa de los elementos: {np.cumsum(vector)}\n")

print(f"Indice del máximo elemento: {np.argmax(vector)}")
print(f"Indice del mínimo elemento: {np.argmin(vector)}\n")

print(f"Vector ordenado: {np.sort(vector)}")
print(f"Elementos únicos: {np.unique(vector)}")
```

## Triángulo de Pascal

```python
n=int(input("Dime el grado:"))
n=n+1
TP = np.zeros((n+1,2*n+1))
TP[0, (2*n+1)//2]=1

for i in range(1,n):
  for j in range(n-i, n+i+1,2):
    TP[i,j] = TP[i-1, j-1]+TP[i-1,j+1]
print('\n'.join([
    ''.join(['{:4}'.format(int(item) if item != 0 else '') for item in row])
    for row in TP
]))
```

### `plt.plot()`
Creación de un gráfico simple

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 16, 24, 32, 40]
y2 = [8, 14, 20, 28, 36]

plt.plot(x, y1, label='Serie 1', color='blue', linestyle='-.')
plt.plot(x, y2, label='Serie 2', color='red', linestyle='--')

plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.title('Gráfico de Líneas Múltiples')
plt.legend()
plt.show()
```

```python
# Strings x valores cuantitativos = Gráfico de barras
categorias = ['A', 'B', 'C', 'D', 'E']
valores = [25, 30, 15, 20, 35]


plt.bar(categorias, valores)
plt.xlabel('Categorías')
plt.ylabel('Valores')
plt.title('Gráfico de Barras')
plt.show()
```

```python
categorias = ['A', 'B', 'C', 'D', 'E']
valores1 = [25, 30, 15, 20, 35]
valores2 = [15, 10, 30, 25, 20]

fig, ax = plt.subplots()
ax.bar(categorias, valores1, label='Serie 1')
ax.bar(categorias, valores2, label='Serie 2', bottom=valores1)

ax.set_xlabel('Categorías')
ax.set_ylabel('Valores')
ax.set_title('Gráfico de Barras Apiladas')
ax.legend()

plt.show()
```

## Fractal Polinomico

```python
# n = número de segmentos
# linspace(vi, área a dividir, vf)
n = int(input("Cantidad de fractales: "))

def polinomio(n):
  x = n*np.cos(np.linspace(0, 2*np.pi, n+1))
  y = n*np.sin(np.linspace(0, 2*np.pi, n+1))
  return x,y

for i in range (n):
  x,y = polinomio(n)

  if (i)%2 == 0:
    plt.fill(x,y, "black")
  else:
    plt.fill(x,y, "white")
  n=n-1
```

[Datos abiertos Colombia](https://www.datos.gov.co)

```python
# Leer desde un archivo CSV
df_colombianos_detenidos = pd.read_csv('https://www.datos.gov.co/api/views/e97j-vuf7/rows.csv?accessType=DOWNLOAD')
df_colombianos_detenidos.head(7) # muestrame las primeras 7 filas
```

```python
# Gráficas estadisticas

select2 = ['PAIS PRISIÓN', 'CANTIDAD']    # Selección de columnas a trabajar
groups2 = ['PAIS PRISIÓN']     #Selección de columnas que agrupará la información

# Creación de nuevo dataframe y le aplicamos la función sum
pais_mayor_detencion = df_colombianos_detenidos[select2].groupby(by = groups2).sum()


# Ordenar el dataframe descendentemente con respecto a la columna 'CANTIDAD'
pais_mayor_detencion = pais_mayor_detencion.sort_values(by = 'CANTIDAD', ascending = False)

pais_mayor_detencion.reset_index(inplace = True)    #Creación de índices numéricos

pais_mayor_detencion.head(7)    #Se imprime el dataframe para visualizar los datos
```

```python
# Crear un diccionario de mapeo de correcciones
correcciones = {'ESPA�A': 'ESPAÑA'}

# Aplicar las correcciones a la columna 'Nombre'
pais_mayor_detencion['PAIS PRISIÓN'] = pais_mayor_detencion['PAIS PRISIÓN'].replace(correcciones)

# Usar groupby para agrupar por la columna 'Pais' y luego sumar los valores
pais_mayor_detencion = pais_mayor_detencion.groupby('PAIS PRISIÓN')['CANTIDAD'].sum().reset_index()
```

```python
# Ordenar el dataframe descendentemente con respecto a la columna 'CANTIDAD'
pais_mayor_detencion = pais_mayor_detencion.sort_values('CANTIDAD', ascending = False)

pais_mayor_detencion.head(7)    #Se imprime el dataframe para visualizar los datos
```

```python
# Limpieza de datos

# Eliminar filas con un valor específico (en este caso, "DESCONOCIDO" en la columna 'PAIS PRISION')
df_filtrado = pais_mayor_detencion[pais_mayor_detencion['PAIS PRISIÓN'] != 'DESCONOCIDO']
df_filtrado.head(7)
```

```python
# Gráfico de Barras
plt.figure(figsize=(10,5))
sns.barplot(data = df_filtrado.head(5), x = "PAIS PRISIÓN", y = "CANTIDAD");
plt.ylabel('Número de presos');
plt.xlabel('País');
```

```python
# Gráfico de barras con más detalles
plt.figure(figsize=(10, 5))  # Aumentar el tamaño de la figura

# Utilizar colores personalizados
colors = sns.color_palette("coolwarm", len(df_filtrado))

# Generar el gráfico de barras con anotaciones
barplot = sns.barplot(data=df_filtrado.head(5), x="PAIS PRISIÓN", y="CANTIDAD", palette=colors)
plt.ylabel('Número de presos', fontsize=12)  # Etiqueta en el eje y
plt.xlabel('País', fontsize=12)  # Etiqueta en el eje x
plt.title('Top 5 de relación del número de presos con el país donde están recluidos', fontsize=14)  # Título del gráfico

# Rotar las etiquetas del eje x para que sean más legibles
plt.xticks(rotation=45, ha="right")



# Agregar etiquetas de datos encima de las barras
"""
for p in barplot.patches: Recorre cada una de las barras en el gráfico de barras (patch)

barplot.annotate(format(p.get_height(), '.0f'), ...: El método annotate agrega una etiqueta de datos a cada barra.

format(p.get_height(), '.0f'): Formatea la altura de la barra como un número entero sin decimales.

(p.get_x() + p.get_width() / 2., p.get_height()): Especifica la posición donde se colocará la etiqueta de datos.

ha='center' y va='baseline': Estos argumentos controlan la alineación horizontal y vertical  de la etiqueta.

xytext=(0, 5): Especifica un desplazamiento (offset) para la posición de la etiqueta.

textcoords='offset points': Indica que el desplazamiento especificado en xytext se interpreta como puntos.
"""
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'baseline',
                     fontsize=10, color='black', xytext = (0, 5),
                     textcoords = 'offset points')

plt.show()
```
