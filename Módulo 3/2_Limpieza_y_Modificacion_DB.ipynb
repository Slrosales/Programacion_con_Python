## Mapeo de correcciones

```python
# Gráficas estadisticas

select2 = ['PAIS PRISIÓN', 'CANTIDAD']    # Selección de columnas a trabajar
groups2 = ['PAIS PRISIÓN']     #Selección de columnas que agrupará la información

# Creación de nuevo dataframe y le aplicamos la función sum
pais_mayor_detencion = df_colombianos_detenidos[select2].groupby(by = groups2).sum()


# Ordenar el dataframe descendentemente con respecto a la columna 'CANTIDAD'
pais_mayor_detencion = pais_mayor_detencion.sort_values(by = 'CANTIDAD', ascending = False)

pais_mayor_detencion.reset_index(inplace = True)    #Creación de índices numéricos

pais_mayor_detencion.head(7)    #Se imprime el dataframe para visualizar los datos
```

```python
# Crear un diccionario de mapeo de correcciones
correcciones = {'ESPA�A': 'ESPAÑA'}

# Aplicar las correcciones a la columna 'Nombre'
pais_mayor_detencion['PAIS PRISIÓN'] = pais_mayor_detencion['PAIS PRISIÓN'].replace(correcciones)

# Usar groupby para agrupar por la columna 'Pais' y luego sumar los valores
pais_mayor_detencion = pais_mayor_detencion.groupby('PAIS PRISIÓN')['CANTIDAD'].sum().reset_index()
```

```python
# Ordenar el dataframe descendentemente con respecto a la columna 'CANTIDAD'
pais_mayor_detencion = pais_mayor_detencion.sort_values('CANTIDAD', ascending = False)
pais_mayor_detencion.reset_index(inplace = True)
pais_mayor_detencion.head(7)    #Se imprime el dataframe para visualizar los datos
```

```python
# Limpieza de datos

# Eliminar filas con un valor específico (en este caso, "DESCONOCIDO" en la columna 'PAIS PRISION')
df_filtrado = pais_mayor_detencion[pais_mayor_detencion['PAIS PRISIÓN'] != 'DESCONOCIDO']
df_filtrado.head(7)
```

```python
# Gráfico de Barras
plt.figure(figsize=(10,5))
sns.barplot(data = df_filtrado.head(5), x = "PAIS PRISIÓN", y = "CANTIDAD");
plt.ylabel('Número de presos');
plt.xlabel('País');
```

```python
# Gráfico de barras con más detalles
plt.figure(figsize=(10, 5))  # Aumentar el tamaño de la figura

# Utilizar colores personalizados
colors = sns.color_palette("coolwarm", len(df_filtrado))

# Generar el gráfico de barras con anotaciones
barplot = sns.barplot(data=df_filtrado.head(5), x="PAIS PRISIÓN", y="CANTIDAD", palette=colors)
plt.ylabel('Número de presos', fontsize=12)  # Etiqueta en el eje y
plt.xlabel('País', fontsize=12)  # Etiqueta en el eje x
plt.title('Top 5 de relación del número de presos con el país donde están recluidos', fontsize=14)  # Título del gráfico

# Rotar las etiquetas del eje x para que sean más legibles
plt.xticks(rotation=45, ha="right")



# Agregar etiquetas de datos encima de las barras
"""
for p in barplot.patches: Recorre cada una de las barras en el gráfico de barras (patch)

barplot.annotate(format(p.get_height(), '.0f'), ...: El método annotate agrega una etiqueta de datos a cada barra.

format(p.get_height(), '.0f'): Formatea la altura de la barra como un número entero sin decimales.

(p.get_x() + p.get_width() / 2., p.get_height()): Especifica la posición donde se colocará la etiqueta de datos.

ha='center' y va='baseline': Estos argumentos controlan la alineación horizontal y vertical  de la etiqueta.

xytext=(0, 5): Especifica un desplazamiento (offset) para la posición de la etiqueta.

textcoords='offset points': Indica que el desplazamiento especificado en xytext se interpreta como puntos.
"""
for p in barplot.patches:
    barplot.annotate(format(p.get_height(), '.0f'),
                     (p.get_x() + p.get_width() / 2., p.get_height()),
                     ha = 'center', va = 'baseline',
                     fontsize=10, color='black', xytext = (0, 5),
                     textcoords = 'offset points')

plt.show()
```

## Agregar columna al DF

La función `numpy.where(condción , x, y)` se utiliza para realizar operaciones condicionales en arrays de NumPy y devuelve índices o valores basados en una condición dada.

Parámetros:

1. **Condición:** Es la condición que se debe evaluar. Para multiples condiciones, se usa & o \|

2. **x:** Es el valor que se asignará a los elementos que cumplan con la condición.

3. **y (opcional):** Es el valor que se asignará a los elementos que no cumplan con la condición. Es opcional y se utiliza cuando se va a asignar un valor diferente a los elementos que no cumplen con la condición o por lo contrario, se rellenarán con valores nulos (NaN) por defecto.



```python
# Creamos una copia del df y a la almacenamos en una nueva variable
df_colombianos_detenidos_alerta = df_colombianos_detenidos.copy()


df_colombianos_detenidos_alerta['ALERTA'] = np.where(
    (df_colombianos_detenidos_alerta['CANTIDAD'] >= 10) & (df_colombianos_detenidos_alerta['SITUACIÓN JURÍDICA'] == 'EN INVESTIGACIÓN'),
    'SI', 'NO'
)

df_colombianos_detenidos_alerta.head()
```

```python
df_colombianos_detenidos_alerta_SI = df_colombianos_detenidos_alerta[df_colombianos_detenidos_alerta['ALERTA'] == 'SI'].reset_index(drop=True)

# Eliminación de una columna
# df_colombianos_detenidos_alerta_SI = df_colombianos_detenidos_alerta_SI.drop('index', axis=1)

df_colombianos_detenidos_alerta_SI.head()
```

```python
select2 = ['ALERTA', 'CANTIDAD']    # Selección de columnas a trabajar
groups2 = ['ALERTA']     #Selección de columnas que agrupará la información

# Creación de nuevo dataframe y le aplicamos la función sum
df_colombianos_detenidos_alerta = df_colombianos_detenidos_alerta[select2].groupby(by = groups2).sum()


# Ordenar el dataframe descendentemente con respecto a la columna 'CANTIDAD'
df_colombianos_detenidos_alerta = df_colombianos_detenidos_alerta.sort_values(by = 'CANTIDAD', ascending = False)

df_colombianos_detenidos_alerta.reset_index(inplace = True)    #Creación de índices numéricos

df_colombianos_detenidos_alerta.head(7)    #Se imprime el dataframe para visualizar los datos
```

## Modificación de Strings

```python
# Creación de un dataframe

# Información que tendrá
data = {'Nombre': ['Luna', 'Giuseppe', 'Laura', 'Rubens'],
        'Edad': ['25(4)', '7(/)', 'hola', '40='],
        'Ciudad': ['Medellín', 'Barranquilla', 'Manizales', 'Barranquilla']}

ejemplo = pd.DataFrame(data)
ejemplo
```

```python
# Limpiar la columna 'Edad' porque tiene caracteres que no nos interesan

"""
Eliminar caracteres no numéricos activandi 'regex=True' y diciendole que
debe eliminar todos los caracteres que no sean dígitos numéricos r'[^0-9]'
y remplazarlos por '' (nada)
"""
ejemplo['Edad'] = ejemplo['Edad'].str.replace(r'[^0-9]', '', regex=True)

"""
Esos strings se convierten a númericos con ayuda de pd.to_numeric
y al errors='coerce', nos aseguramos que aquellos strings que no tienen
conversión a número, se registren como NaN (Not a Number)
"""
ejemplo['Edad'] = pd.to_numeric(ejemplo['Edad'], errors='coerce')

ejemplo
```
