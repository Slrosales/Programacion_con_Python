# Manejo de Archivos

Python proporciona funciones básicas y métodos necesarios para manipular archivos, como: leer, escribir, agregar contenido y cerrar archivos.

**Propósito:**
- Persistencia de datos: Guardar y recuperar información entre ejecuciones de un programa.
- Manipulación de información: Procesar grandes cantidades de texto o binarios.
- Comunicación con otros programas: Importar o exportar datos en formatos estándar como CSV, JSON, XML, etc.

etc.

**Estructura**

`open(nombre_del_archivo, modo)`

**nombre_del_archivo:** puedes poner el nombre del archivo si lo tienes en el mismo directorio de tu entorno o poner el path hasta el archivo

**modo**: Es una cadena que especifica el modo en el que se abrirá el archivo.

Como:

- **'r'**:
  - Modo de lectura.
  - Si el archivo no existe, lanza una excepción.
  - Es el modo por defecto.
- **'a'**:
  - Modo de anexado (append).
  - Si el archivo no existe, lo crea.
  - Si existe, anexa el último dato al final
- **'w'**:
  - Modo de escritura.
  - Abre el archivo para escritura, crea el archivo si no existe o trunca el archivo si ya existe.
- **'x'**:
  - Modo de creación exclusiva.
  - Abre el archivo para escritura exclusiva, fallando si el archivo ya existe.
- **'t'**:
  - Es el modo por defecto.
  - Modo texto.
  - Se utiliza para abrir el archivo en modo texto.
- **'b'**:
  - Modo binario.
  - Se utiliza para abrir el archivo en modo binario, por ejemplo, imágenes.

Se puede combinar modos. Por ejemplo, `'rb'` abriría el archivo para lectura en modo binario, mientras que `'w+'` abriría el archivo para escritura y lectura, truncando el archivo si ya existe.

**Crear un archivo y escribir en un archivo**

```python
with open("archivo.txt", "w") as archivo:
  archivo.write("¡Hola mundo!")
```

**Leer de un archivo**

```python
# Accediendo al archivo a través de un path
with open("/content/archivo.txt", "r") as archivo:
  contenido = archivo.read()
  print(contenido)
```

**Agregar contenido a un archivo existente**

```python
with open("archivo.txt", "a") as archivo:
  archivo.write("\n¿Cómo estas?")
```

## Ejemplo 1

```python
with open("archivo.txt", "r") as archivo:
    lineas = archivo.readlines()
    for linea in lineas:
        print(linea.strip())
```

### `.strip()`

En el contexto de leer líneas de un archivo, se usa para remover posibles saltos de línea `\n` que pueden estar presentes al final de las líneas.

En general, sirven para eliminar espacios en blanco (u otros caracteres, si se especifican) del inicio y del final de la cadena.

```python
texto = "###¡Hola, Mundo!###"
print(texto.strip('#'))  # Salida: "Hola, Mundo!"
```

```python
print(texto.replace('#', ','))
```

## Ejemplo 2

```python
def agregar_registro(archivo):
  print("\nIngresa los valores. Escribe 'salir' cuando quieras salir.")

  with open(archivo, "a") as f:
    # Se ingresan valores hasta que el usuario quiera salir
    while True:
      valor = input()
      if valor.lower() == "salir":
          break
      f.write(valor + "\n")
```

```python
def ver_contenido(archivo):
  with open(archivo, "r") as f:
    contenido = f.read()
    print(contenido)
```

```python
def acceder_linea(archivo, num_linea):
  with open(archivo, "r") as f:
    lineas = f.readlines()
    try:
        print(lineas[num_linea - 1].strip())  # -1 porque las listas inician en 0
    except IndexError:
        print("El número de línea está fuera de rango.")
```

```python
archivo = input("Escriba el nombre del archivo: ")
archivo = archivo + ".txt" # .csv xml .json
archivo
```

```python

while True:
  print("\nMenú:")
  print("1. Agregar valores")
  print("2. Ver contenido del archivo")
  print("3. Ver contenido de una línea específica")
  print("4. Salir")

  res = input("Selecciona una opción: ")

  if res == "1":
      agregar_registro(archivo)
  elif res == "2":
      ver_contenido(archivo)
  elif res == "3":
      num_linea = int(input("Introduce el número de línea al que deseas ver: "))
      acceder_linea(archivo, num_linea)
  elif res == "4":
      break
  else:
      print("Opción no válida.")
```
